```python
# Testat: Übung 1 A 

# Liste A von positiven natürlichen Zahlen 
# Ziel: Es soll die Zahl ausgeben werden, die 
# die meisten Einsen in der Binärdarstellung enthält

#listOfNums = [14,1,25,38,69,22,70,1995]
listOfNums = [27,29,30,31,32,34,60]
numWithMostOnes = 0
mostOnes = 0

def dezToBin(n):
    if n == 0:
        return [0]
    result = []
    while n > 0:
        result.append(n%2)
        n = n // 2
    return list(reversed(result))


for z in listOfNums:
    binZ = dezToBin(z)
    addedOnes = 0 # zählt die Anzahl der Einsen

    # Nimmt alle Ziffern aus der Liste und addiert sie miteinander
    # Nullen spielen keine Rolle
    
    for p in binZ: # alternativ mit einer While und pop() lösbar
        addedOnes = addedOnes + p

    if (mostOnes < addedOnes):
        mostOnes = addedOnes
        numWithMostOnes = z
    
print("Meiste Einsen in der Binärdarstellung (", mostOnes, ") in Zahl: ",numWithMostOnes)
```

```python
# WS Übung 02 A

# Liste mit beliebig vielen Zahlen.... davon ggT berechnen

import math

def berechneGGT(nums):
    
    nums.reverse() # Einfach zur besseren Übersicht...
    
    while(len(nums) > 1):
        tempGCD = math.gcd(nums.pop(), nums.pop())
        print("Momentaner GGT: "+str(tempGCD))
        nums.append(tempGCD)
        #Alternativ while zur if machen und kursive Funktion: berechneGGT(nums)
    else:
        print("GGT: " +str(nums.pop()))
        
berechneGGT([7,49,77,70,8])

#########
'''
# Aufgabe 2

# Restklassenring ... eine Zahl innerhalb des Restklassenrings soll auf Nilpotenz überprüft werden
# Nilpotent heißt: Zahl**Hochzahl == 0
# Liste zwischen 2 und 100 (Modulo)
# Ergebnis: Wir wollen wissen welche Elemente der Liste neben 0 noch weitere Nilpotenzen besitzen

0*0, 0*1, 0*2, 0*3, 0*4
1*0, 1*1, 1*2, 1*3, 1*4
2*0, ...
3..
4..'''

listOfNums = []
listOfNils = []

# Liste mit Zahlen befüllen
for z in range (2,101):
    listOfNums.append(z)
    
listOfNums.reverse()

# Eigentliche Funktion, die stückweise mit einer Zahl arbeitet
def nilpotenzSucher(mod):
    for a in range(1,mod): # Wir fangen bei 1 an, weil 0 nilpotent ist und eine beliebe andere Zahl hoch 0, gleich 1 ist.
        for n in range(1,mod):
            if ((a**n) % mod == 0):
                return True # Das reicht in diesem Fall, weil wir lediglich überprüfen wollen, ob mindestens eine Nilpotenz neben der Null existiert
            # Alternativ: ...
                #print("Nilpotenz gefunden in "+str(mod));


while(len(listOfNums) > 0):
    mod = listOfNums.pop()
    if(nilpotenzSucher(mod)):
        listOfNils.append(mod)

print("Nilpotenzen gefunden in Mod: ",listOfNils)
print("Alle gefunden Mods sind durch 2 oder durch 3 teilbar... siehe komisches Dokument.")
```

```python
# Übung 2B WS

import math
#import numpy as n thereotisch möglich, geht aber aufgrund der aktuellen Version irgendwie nicht... mit lcm(a,b)

def lgcdLcm(a,b):
    results = []
    
    ggT = math.gcd(a,b)
    kgV = a*b // math.gcd(a,b)
    produkt = ggT * kgV
    
    results.append(ggT)
    results.append(kgV)
    results.append(produkt)
    
    return results

lgcdLcm(2,4)


#Aufgabe 2:
#zwei Listen um Brüche darzustellen
#es sollen der Hauptnenner, der addierte Zähler und der gekürzte Nenner ausgegeben werden


bruch1 = [3,7]
bruch2 = [6,3]

# bruch1[0] = 3 | bruch1[1] = 7
# bruch2[0] = 6 | bruch2[1] = 2

def maindenomiator(bruch1, bruch2):
    listOfResults = []
    
    hauptnenner = bruch1[1] * bruch2[1]
    
    bruch1[0] = bruch1[0] * bruch2[1]
    bruch2[0] = bruch2[0] * bruch1[1]
    
    bruch1[1] = hauptnenner
    bruch2[1] = hauptnenner
    
    print(bruch1, bruch2)
    
    
    addZaehler = bruch1[0] + bruch2[0]
    
    gcd = math.gcd(addZaehler, hauptnenner)
    if(gcd != 1):
        gekuerzterNenner = hauptnenner//gcd
        addZaehler = addZaehler // gcd
    else:
        gekuerzterNenner = hauptnenner
                   
    listOfResults.append(hauptnenner)
    listOfResults.append(addZaehler)
    listOfResults.append(gekuerzterNenner)
                   
    return listOfResults


maindenomiator(bruch1,bruch2)
```

```python
# WS Übung 3 A

from sympy import isprime

primList = [] # Speichert alle n-Zahlen, die mindestens 42 Primzahlen innerhalb der gegebenen Range erhalten

for n in range(1,300):
    primCounter = 0
    
    for i in range(n**2, (n+1)**2):
        if(isprime(i)):
            primCounter = primCounter +1
        
    if(primCounter == 42):
        primList.append((n))
    
primList
```
