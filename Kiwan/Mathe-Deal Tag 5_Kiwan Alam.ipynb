```python
#Aufgabe 1

#step-Fkt:
    #Variable x
    #solange nicht gilt: x*m1*m2 % m3 == a
        #x um 1 erhöhen
    #Ergebnis aus x*m1*m2 zurückgeben
    
    
#crt-Fkt:
    #3 Werte s1, s2, s3 mithilfe von step() berechnen
    #s-Werte addieren
    #Moduln multiplizieren für die Schrittweite
    #solange s größer als m ist
        #m von s abziehen
        #so ist s die kleinste positive Lösung des Kongruenzsystems
    #s zurückgeben
```

```python
def step (m1, m2, m3, a):
    x = 0
    while (x * (m1 * m2)) % m3 != a:
        x += 1
    return x * (m1 * m2)

step(3, 4, 17, 16)
```

```python
def crt (a1, m1, a2, m2, a3, m3):
    s1 = step(m2, m3, m1, a1)
    s2 = step(m1, m3, m2, a2)
    s3 = step(m1, m2, m3, a3)
    s = s1 + s2 + s3
    m = m1 * m2 * m3
    while s > m:
        s -= m
    return s

crt (1, 3, 1, 4, 16, 17)
#perfekt :)
```

```python
from math import log

def near2Prime(n1, n2, p):
    differenz1 = abs(n1 - p)
    differenz2 = abs(n2 - p)
    if differenz1 < differenz2:
        return n1
    elif differenz2 < differenz1:
        return n2
    else:
        print("Oh, sind beide wohl etwa gleich weit entfernt")
    
near2Prime(2*10**18, 4*10**18, (10**16 * log(10**16)))
```
